#Классы

```python
it_specialists =[('jon','de', 150),
                 ('carl','da', 250),
                 ('gleb','ds', 350)]
```

```python
class ItSpecialist:
  def __init__(self,name,speciality,salary):
    self.name=name
    self.speciality=speciality
    self.salary=salary
  def tell_about_yourself(self):
    return f'Hi! My name is {self.name}, I am a {self.speciality}, my sal is {self.salary}'
```

```python
jon = ItSpecialist('jon','de', 150)
```

```python
jon
```

```python
jon.name
```

```python
jon.speciality
```

```python
jon.salary
```

```python
carl = ItSpecialist('carl','da', 250)
```

```python
jon.tell_about_yourself()
```

```python
carl.tell_about_yourself()
```

```python
class DataScientist(ItSpecialist):
  def __init__(self, name, salary):
    super().__init__(name, 'ds', 350)
  def tell_about_yourself(self):
    return f'Hi! My name is {self.name}, I am a {self.speciality}, my sal is {self.salary}k per second'
  def ml(self, data):
    return 'profit'
```

```python
oleg=DataScientist('oleg',350)
```

```python
oleg.ml(1)
```

```python
oleg.tell_about_yourself()
```
